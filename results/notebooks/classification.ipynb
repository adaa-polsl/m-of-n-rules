## Comparison with Decision Tree

```python
Tables.configure('../classification/decision_tree')
decision_tree_metrics = pd.concat(Tables.query('*', 'metrics', as_pandas=True))
columns: list = decision_tree_metrics.columns.tolist()
columns.remove('dataset')
columns.insert(0, 'dataset')
decision_tree_metrics = decision_tree_metrics[columns]

decision_tree_metrics = decision_tree_metrics.set_index('dataset')
rules_metrics = metrics.set_index('dataset')


pd.DataFrame({
    'rules BAcc (train)': rules_metrics['BAcc (train)'],
    'model_type': rules_metrics['model_type'],
    'DC BAcc (train)': decision_tree_metrics['BAcc (train)'],
    'rules BAcc (test)': rules_metrics['BAcc (test)'],
    'DC BAcc (test)': decision_tree_metrics['BAcc (test)'],
}, index=rules_metrics.index)
```

Aggregated results

```python
agg_metrics = pd.DataFrame({
    'rules BAcc (train)': rules_metrics['BAcc (train)'],
    'model_type': rules_metrics['model_type'],
    'rules BAcc (test)': rules_metrics['BAcc (test)'],
    'rules': rules_metrics['rules']
}, index=rules_metrics.index).groupby('model_type').agg(string_supporting_mean)

agg_metrics['DC BAcc (train)'] = decision_tree_metrics['BAcc (train)'].mean()
agg_metrics['DC BAcc (test)'] = decision_tree_metrics['BAcc (test)'].mean()
```

```python
agg_metrics = pd.DataFrame({
    'rules BAcc (train)': rules_metrics['BAcc (train)'],
    'model_type': rules_metrics['model_type'],
    'rules BAcc (test)': rules_metrics['BAcc (test)'],
}, index=rules_metrics.index).groupby('model_type').agg(string_supporting_mean)

agg_metrics['DC BAcc (train)'] = decision_tree_metrics['BAcc (train)'].mean()
agg_metrics['DC BAcc (train) (std)'] = decision_tree_metrics['BAcc (train)'].std()
agg_metrics['DC BAcc (test)'] = decision_tree_metrics['BAcc (test)'].mean()
agg_metrics['DC BAcc (test) (std)'] = decision_tree_metrics['BAcc (test)'].std()

agg_metrics
```
