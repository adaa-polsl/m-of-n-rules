```python
import pandas as pd
from utils import string_supporting_mean, print_rules
from experiments_utils.results.tables import Tables

Tables.configure('../classification/min_supp_20')

metrics = pd.concat(Tables.query('*', '*', '*', 'metrics', as_pandas=True))
columns = [c for c in metrics.columns.tolist() if 'Unnamed' not in c]
columns.remove('model_type')
columns.remove('dataset')
metrics = metrics[['dataset', 'model_type'] + columns]
metrics
```

```python
res = []
for datasets in metrics['dataset'].unique():
    m_of_n_bacc = -1
    m_of_n_bacc = metrics[
        (metrics['dataset'] == datasets) & (metrics['model_type'] == 'at_least_M-of-N')
    ]['BAcc (test)'].iloc[0]
    tmp = metrics[
        (metrics['dataset'] == datasets) & (metrics['model_type'] == 'exact_M-of-N')
    ]['BAcc (test)'].iloc[0]
    if tmp > m_of_n_bacc:
        m_of_n_bacc = tmp
    tmp = metrics[
        (metrics['dataset'] == datasets) & (metrics['model_type'] == 'complex')
    ]['BAcc (test)'].iloc[0]
    if tmp > m_of_n_bacc:
        m_of_n_bacc = tmp
    bacc_plain = metrics[
        (metrics['dataset'] == datasets) & (metrics['model_type'] == 'plain')
    ]['BAcc (test)'].iloc[0]

    diff = m_of_n_bacc - bacc_plain
    res.append((datasets, m_of_n_bacc, bacc_plain, diff))

sorted(res, key=lambda x: x[3], reverse=True)
```

```python
from glob import glob
import os
folds_results = []
variants = ['at_least_M-of-N', 'complex', 'exact_M-of-N', 'plain']

best_fold_number = None
best_bacc = -1
for variant in variants:
    for path in glob('../classification/min_supp_20/iris/no_discretization/exact_M-of-N/cv/*'):
        fold_number = os.path.basename(path)
        bacc_plain = pd.read_csv(f'../classification/min_supp_20/iris/no_discretization/plain/cv/{fold_number}/metrics.csv')['rules'].iloc[0]
        bacc = bacc_plain - pd.read_csv(f'{path}/metrics.csv')['rules'].iloc[0]
        if bacc >= best_bacc:
            best_bacc = bacc
            best_fold_number = fold_number

print(best_bacc, best_fold_number)
        
```

```python
with open(f'../classification/min_supp_20/iris/no_discretization/at_least_M-of-N/cv/{best_fold_number}/rules/rules.txt', 'r') as f:
    rules = f.read()
    new_rules = []
    start = False
    for rule in rules.split('\n'):
        if start and rule.strip() != '':
            rules_str, cov_str = rule.split('(p=')
            new_rules.append(f'{rules_str} (p={cov_str.replace(".0", "")} \\\\')
        if '____' in rule:
            start = True
for r in new_rules:
    print(r)
pd.read_csv(f'../classification/min_supp_20/iris/no_discretization/at_least_M-of-N/cv/{best_fold_number}/metrics.csv')['BAcc (test)']
```
