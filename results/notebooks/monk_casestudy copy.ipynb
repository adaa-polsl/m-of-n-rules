```python
with open(f'../classification/min_supp_20/monk-1/no_discretization/at_least_M-of-N/rules/rules.txt', 'r') as f:
    rules = f.read()
    new_rules = []
    start = False
    for rule in rules.split('\n'):
        if start and rule.strip() != '':
            rules_str, cov_str = rule.split('(p=')
            new_rules.append(f'{rules_str} (p={cov_str.replace(".0", "")} \\\\')
        if '____' in rule:
            start = True
for r in new_rules:
    print(r)
pd.read_csv(f'../classification/min_supp_20/monk-1/no_discretization/at_least_M-of-N/metrics.csv')['BAcc (test)']
```
