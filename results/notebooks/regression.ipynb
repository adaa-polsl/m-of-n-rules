## Comparison with Decision Tree

```python
Tables.configure('../regression/decision_tree')
decision_tree_metrics = pd.concat(Tables.query('*', 'metrics', as_pandas=True))
columns: list = decision_tree_metrics.columns.tolist()
columns.remove('dataset')
columns.insert(0, 'dataset')
decision_tree_metrics = decision_tree_metrics[columns]

decision_tree_metrics = decision_tree_metrics.set_index('dataset')
rules_metrics = metrics.set_index('dataset')


pd.DataFrame({
    'rules rRMSE (train)': rules_metrics['rRMSE (train)'],
    'model_type': rules_metrics['model_type'],
    'DC rRMSE (train)': decision_tree_metrics['rRMSE (train)'],
    'rules rRMSE (test)': rules_metrics['rRMSE (test)'],
    'DC rRMSE (test)': decision_tree_metrics['rRMSE (test)'],
}, index=rules_metrics.index)
```

```python
agg_metrics = pd.DataFrame({
    'rules rRMSE (train)': rules_metrics['rRMSE (train)'],
    'model_type': rules_metrics['model_type'],
    'rules rRMSE (test)': rules_metrics['rRMSE (test)'],
}, index=rules_metrics.index).groupby('model_type').agg(string_supporting_mean)

agg_metrics['DC rRMSE (train)'] = decision_tree_metrics['rRMSE (train)'].mean()
agg_metrics['DC rRMSE (test)'] = decision_tree_metrics['rRMSE (test)'].mean()

agg_metrics
```

```python
agg_metrics = pd.DataFrame({
    'rules rRMSE (train)': rules_metrics['rRMSE (train)'],
    'model_type': rules_metrics['model_type'],
    'rules rRMSE (test)': rules_metrics['rRMSE (test)'],
}, index=rules_metrics.index).groupby('model_type').agg(string_supporting_mean)

agg_metrics['DC rRMSE (train)'] = decision_tree_metrics['rRMSE (train)'].mean()
agg_metrics['DC rRMSE (train) (std)'] = decision_tree_metrics['rRMSE (train)'].std()
agg_metrics['DC rRMSE (test)'] = decision_tree_metrics['rRMSE (test)'].mean()
agg_metrics['DC rRMSE (test) (std)'] = decision_tree_metrics['rRMSE (test)'].std()

agg_metrics
```
